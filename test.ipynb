```python
import pandas as pd
from tensorflow.keras.models import load_model
```

```python
df = pd.read_csv('datasets/train.csv')
```

```python
model = load_model('nlp_toxic.keras')
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer()
max_len = 280

texts = df['text'].astype(str).tolist()  # Supondo que 'texto' seja a coluna com os textos nos seus novos dados
sequences = tokenizer.texts_to_sequences(texts)  # Use o mesmo tokenizer que vocÃª usou para treinar o modelo
data = pad_sequences(sequences, maxlen=max_len)  # Use o mesmo max_len que vocÃª usou para treinar o modelo
```

```python
texts
```

```python
# Fazer previsÃµes com o modelo treinado
previsoes = model.predict(data)
```

```python
df_trained = df
df_trained['label'] = previsoes
df_trained
```
